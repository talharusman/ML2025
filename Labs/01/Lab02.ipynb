```python
# Question 02 (A,B,C,D,E)

df = pd.read_csv("cancer patient data sets.csv")

display(df['Level'].unique())
print(df.columns)
print(df.head(50))
print(df.tail(30))
print(df.info())

print(df.shape)


```

```python
# Question 02 (F)


category_counts = df['Level'].value_counts()
print("Category Count:", category_counts)

sns.countplot(data=df, x='Level')
plt.title("Target Variable Distribution")
plt.show()
```

```python
# Task 03 - Combined Graphs

fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# --- Correlation Heatmap ---
data_corr = df.drop(columns=['Patient Id', 'Level'])
sns.heatmap(data_corr.corr(), annot=True, cmap="viridis", linewidths=0.5, ax=axes[0, 0])
axes[0, 0].set_title("Feature Correlation Heatmap", fontsize=12, fontweight="bold")

# --- Age Distribution ---
sns.histplot(df['Age'], bins=15, kde=True, color="skyblue", edgecolor="black", ax=axes[0, 1])
axes[0, 1].set_xlabel("Age")
axes[0, 1].set_ylabel("Frequency")
axes[0, 1].set_title("Age Distribution", fontsize=12, fontweight="bold")

# --- Frequent Cold Count Plot ---
sns.countplot(x='Frequent Cold', data=df, palette="Set2", ax=axes[1, 0])
axes[1, 0].set_xlabel("Frequent Cold (Yes/No)")
axes[1, 0].set_ylabel("Count")
axes[1, 0].set_title("Frequency of Cold Symptoms", fontsize=12, fontweight="bold")

# --- Obesity Boxplot ---
sns.boxplot(y='Obesity', data=df, color="lightcoral", ax=axes[1, 1])
axes[1, 1].set_ylabel("Obesity Level")
axes[1, 1].set_title("Boxplot of Obesity", fontsize=12, fontweight="bold")

# Adjust layout
plt.tight_layout()
plt.show()
```
